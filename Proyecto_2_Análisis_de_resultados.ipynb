# **Revisión de resultados para asignación de Clusters 2D**

```python
import pandas as pd

ruta = "customer_data_cluster_2D.xlsx"
df_2d = pd.read_excel(ruta)
df_2d.head()
```

Para el presente analisis no se consideraran aquellas clusterizaciones menores a 6 o mayores de 11, ni aquellos en los que los valores no pudieres ser debidamente clasificados (presencian de cluster clusters -1)

```python
features_2D = ["frequency", "monetary_value", "avg_ticket", "num_unique_products", "recency"]
clustering_columns_2D = ["Kmeans_mix", "Kmeans_cal", "MS_cal", "MS_dav", "Agg_cal"]
stats_funcs_2D = ["min", "max", "mean", "median","std"]
```

```python
def generar_estadisticos(df, methods, features, stats):
    resultados = {}
    for metodo in methods:
        # Agrupar por cluster y calcular estadísticos
        grouped = df.groupby(metodo)[features].agg(stats)
        # Reorganizar columnas para mejor legibilidad
        grouped.columns = ['_'.join(col).strip() for col in grouped.columns.values]
        resultados[metodo] = grouped
    return resultados
```

```python
resultados = generar_estadisticos(df_2d, clustering_columns_2D, features_2D, stats_funcs_2D)

# Mostrar resultados para cada método
for metodo, stats_df in resultados.items():
    print(f"\n=== Método: {metodo} ===")
    display(stats_df)
```
